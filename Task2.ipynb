### Data cleaning

```python
#Check null values
df.isnull()
```

```python
df.isnull().sum()
```

```python
#Remove 'Cabin row', as it has many null values
df.drop(columns='Cabin', axis = 1, inplace = True)
```

```python
#Remove null values of 'Embarked' columns
df.dropna(subset=['Embarked'], inplace=True)
```

```python
#Fill null values of 'Age' with the median
df['Age'].fillna(df['Age'].median(), inplace = True)
```

```python
df.info()
```

## Data Visualization

```python
#Heatmap
sns.heatmap(df[["Survived", "SibSp", "Parch", "Age", "Fare"]].corr(), annot=True)
```

```python
#Count of survived passengers plot
sns.countplot(x='Survived',data=df,palette='pastel')
plt.title('Count of Survived Passengers')
plt.xlabel('Survived')
plt.ylabel('Count')
```

```python
#Survived Passengers rate by gender plot
sns.catplot(x = 'Sex', hue = 'Survived', kind = 'count', data = df, palette='pastel')
plt.title('Count of Survived Passengers by gender')
plt.xlabel('Survived')
plt.ylabel('Count')
```

```python
#Survived Passengers rate by Fare range plot
plt.figure(figsize = (10, 5))
df['Fare_range'] = pd.qcut(df['Fare'], 5)
colors = sns.color_palette("Set1", n_colors=len(df['Fare_range'].unique()))
sns.barplot(x = 'Fare_range', y = 'Survived', data = df, palette='pastel')
plt.xlabel('Fare range')
plt.ylabel('Survived')
plt.title('Survived Passengers rate by Fare range')

```

```python
#Survived Passengers rate by Class plot
sns.barplot(x = 'Pclass', y = 'Survived', data = df, palette='pastel')
plt.xlabel('Class')
plt.ylabel('Survived')
plt.title('Survived Passengers rate by Class')
```

```python
# Survived Passengers rate by Age plot
plt.figure(figsize = (15, 10)) 
sns.histplot(data=df, x='Age', bins=30, hue='Survived', multiple='stack', palette='pastel', alpha=0.7)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram of Age with Survival Information')
plt.show()
```

```python
# Survived Passengers rate by number of siblings/spouses plot
plt.figure(figsize = (10, 5))
sns.barplot(x = 'SibSp', y = 'Survived', data = df, palette='pastel')
plt.xlabel('# of siblings/spouses')
plt.ylabel('Survived')
plt.title('Survived Passengers rate by Number of siblings/spouses above')
```

```python
# Survived Passengers rate by number of parents/children above plot
plt.figure(figsize = (10, 5))
sns.barplot(x = 'Parch', y = 'Survived', data = df, palette='pastel')
plt.xlabel('# of parents/children')
plt.ylabel('Survived')
plt.title('Survived Passengers rate by Number of parents/children above')
```

```python
# Survived Passengers rate by port of embarkation plot
plt.figure(figsize = (10, 5))
sns.barplot(x = 'Embarked', y = 'Survived', data = df, palette='pastel')
plt.xlabel('Embarked')
plt.ylabel('Survived')
plt.title('Survived Passengers rate by Port of Embarkation')
```

## Conclusions

With this Titanic data analysis, we can conclude the following:
- Survival Rate: Only one-third of the total passengers on board survived.
- Gender Disparity: The survival rate is notably higher among females, with the majority of survivors being women.
- Fare Rank Influence: Passengers with higher fare ranks demonstrated a greater likelihood of survival.
- Class Disparities: First-class passengers had significantly higher chances of surviving compared to lower-class passengers.
- Age and Survival: Although age alone did not directly influence survival, it is noteworthy that the majority of passengers fell within the 25 to 35 years age range.
- Siblings/Spouse Influence: Passengers with only one sibling or spouse on board exhibited a higher survival rate.
- Parents/Children Influence: Those with three parents or children on board had a higher chance of survival.
- Embarkation Point: Passengers who embarked at port C (Cherbourg) experienced a higher survival rate.
